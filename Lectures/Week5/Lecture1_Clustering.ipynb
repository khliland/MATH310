# The standard inner product and norm in ${\bf R}^n$ ([chapter 1.4](https://web.stanford.edu/~boyd/vmls/vmls.pdf#page=29) in VMLS)


<font size="5">  

The the standard [inner product (scalar product)](https://web.stanford.edu/~boyd/vmls/vmls-slides.pdf#page=24) between two vectors 
$${\bf u} = \begin{bmatrix}
 u_1\\
 \vdots\\
u_n
\end{bmatrix}
, {\bf v}= \begin{bmatrix}
 v_1\\
 \vdots\\
v_n
\end{bmatrix} \in {\bf R}^n$$ is defined as the number

$$ {\bf u}\cdot {\bf v}={\bf u}^t{\bf v} =\sum_{i=1}^n u_i v_i.$$

The [<ins>norm</ins>](https://web.stanford.edu/~boyd/vmls/vmls-slides.pdf#page=48) of ${\bf v}$ in ${\bf R}^n$ is defined as

$$\|{\bf v}\|=\sqrt{{\bf v}\cdot{\bf v}} = \sqrt{\sum_{i=1}^n v_i v_i},$$

i.e. the square root of the inner product of ${\bf v}$ with itself.

</font>

# The distance between vectors in Euclidean space ([chapter 3.2](https://web.stanford.edu/~boyd/vmls/vmls.pdf#page=58) in VMLS)

<font size="5">  

The <ins> distance</ins> between two vectors ${\bf u}, {\bf v}\in {\bf R}^n$ is defined as the norm of their difference

$$dist({\bf u}, {\bf v})= \|{\bf u}-{\bf v}\|=\sqrt{({\bf u}-{\bf v})^t({\bf u}-{\bf v})}=\sqrt{\sum_{i=1}^n(u_i-v_i)^2}.$$

</font>

# The Cauchy-Schwartz (CS) inequality ([chapter 3.4](https://web.stanford.edu/~boyd/vmls/vmls.pdf#page=66)  in VMLS)

<font size="5">  

__Theorem ([Cauchy-Schwartz inequality](https://en.wikipedia.org/wiki/Cauchy\%E2\%80\%93Schwarz_inequality))__

If ${\bf u}, {\bf v}\in {\bf R}^n$,  then 

$$|{\bf u}\cdot {\bf v}| \leq \|{\bf u}\|\|{\bf v}\|,$$
   
and equality only applies when either ${\bf u}$ or ${\bf v}$ is equal to ${\bf 0}$, or when ${\bf u}=k{\bf v}$ for $k\in {\bf R}$.


<ins>Proof:</ins>

If ${\bf u}$ or ${\bf v}$ is equal to ${\bf 0}$ equality in the above formula holds by inspection. Therefore we assume that ${\bf u},{\bf v}\neq {\bf 0}$. Let ${\bf p}={\bf v}({\bf v}^t{\bf v})^{-1}{\bf v}^t{\bf u}=(\frac{{\bf v}\cdot{\bf u}}{{\bf v}\cdot{\bf v}}){\bf v}$ be the orthogonal projection of ${\bf u}$ onto ${\bf v}$. Because ${\bf p}$ is orthogonal to the residual vector $({\bf u}-{\bf p})$, [Pythagoras theorem](https://en.wikipedia.org/wiki/Pythagorean_theorem) implies 

$$\|{\bf p}\|^2+\|{\bf u}-{\bf p}\|^2=\|{\bf u}\|^2\Rightarrow \|{\bf p}\|^2\leq \|{\bf u}\|^2.$$ 

Because ${\bf p}=(\frac{{\bf v}\cdot{\bf u}}{{\bf v}\cdot{\bf v}}){\bf v}=\frac{({\bf v}\cdot{\bf u})}{\|{\bf v}\|^2}{\bf v}$,
$$\|{\bf p}\|^2={\bf p}\cdot {\bf p}=\frac{({\bf v}\cdot{\bf u})^2}{\|{\bf v}\|^4}\|{\bf v}\|^2=\frac{({\bf v}\cdot{\bf u})^2}{\|{\bf v}\|^2}\leq\|{\bf u}\|^2$$ $$\Downarrow$$  $$({\bf v}\cdot{\bf u})^2\leq\|{\bf u}\|^2\|{\bf v}\|^2,$$ 
   
and the CS inequality follows by taking the square root of both sides of this inequality $\blacksquare$ 

An alternative proof of CS is given [on page 57 in VMLS](https://web.stanford.edu/~boyd/vmls/vmls.pdf#page=67).
    
</font>  

# Definition: Angle between vectors in ${\bf R}^n$

<font size="5">  

We define the cosine of the angle $\theta$ between two vectors  ${\bf u}, {\bf v}\in {\bf R}^n$ by the formula
$$\cos(\theta)=\frac{{\bf u}\cdot {\bf v}}{\|{\bf u}\|\|{\bf v}\|}.$$
  
  
This definition is sound because of the CS-inequality: 
$$|{\bf u}\cdot{\bf v}|\leq\|{\bf u}\|\|{\bf v}\|\Leftrightarrow \frac{|{\bf u}\cdot{\bf v}|}{\|{\bf u}\|\|{\bf v}\|} \leq 1\Leftrightarrow -1\leq \frac{{\bf u}\cdot {\bf v}}{\|{\bf u}\|\|{\bf v}\|}\leq 1 \Rightarrow -1\leq \cos(\theta)\leq 1.$$ 
  
Finally we define the angle $\theta$ between  ${\bf u}$ and ${\bf v}$ as
$$\theta = \arccos\left (\frac{{\bf u}\cdot {\bf v}}{\|{\bf u}\|\|{\bf v}\|}\right)$$
where $\arccos$ denotes the inverse cosine, normalized to lie in the interval $[0, \pi]$.  

</font> 

# The triangle inequality

<font size="5"> 
    
__Theorem ([the triangle inequality](https://en.wikipedia.org/wiki/Triangle_inequality))__ 

For all ${\bf u}, {\bf v}\in {\bf R}^n$ the following inequality holds:
$$\|{\bf u}+{\bf v}\|\leq\|{\bf u}\|+\|{\bf v}\|$$
<ins>Proof:</ins>

$$\|{\bf u}+{\bf v}\|^2 = ({\bf u}+{\bf v})\cdot ({\bf u}+{\bf v})={\bf u}\cdot{\bf u} +2{\bf u}\cdot {\bf v}+{\bf v}\cdot {\bf v}$$ $$\leq \|{\bf u}\|^2+2|{\bf v}\cdot {\bf u}|+\|{\bf v}\|^2$$  $$\textrm{(use CS here)}$$ $$\leq \|{\bf u}\|^2+2\|{\bf u}\|\|{\bf v}\|+\|{\bf v}\|^2$$ $$= (\|{\bf u}\|+\|{\bf v}\|)^2.$$

The 2nd inequality holds due to the Cauchy-Schwartz inequality. The result follows by taking the square root of both sides of this inequality$\blacksquare$
    
</font>

# Examples with simulated data (Julia code)

<font size="5">

In Julia we can do the k-means clustering either by using the [VMLS-libray function  *kmeans*](https://github.com/VMLS-book/VMLS.jl/blob/master/src/VMLS.jl), or by implementing our own version of the k-means algorithm,  see *mykmeans* below.
    
</font>    

## Lets start by generating a random "artificial" dataset X containing 3 point-clouds in 2 dimensions:

```julia
using Random
Random.seed!(1234) # This line assures the same random dataset to be generated each time.

nn = 100;     # Generate 3 random "clouds" of datapoints (samples), each of size nn.
X = [randn((nn,2))*0.7 .+ [ 1  1];
    randn((nn,2))*0.7  .+ [-3 -5];
    randn((nn,2))*0.7  .+ [-4 -2]];
# Each row of X corresponds to a datapoint
```

### A scatterplot shows the generated dataset:

```julia
using Plots #b Precompiles on every startup (~20 secondss)
gr() # Needs modules Plots and GR to be installed, may need a rebuild of GR with ']build GR'
default(size=(600, 400), fmt = :png) # Default plot size, change output format to png
```

```julia
# Define and display a plot of the raw random data
sp = scatter(X[:,1],X[:,2], title = "Random generated pointcloud 2D-data", legend = false)
display(sp)
```

### Apply the k-means clustering algorithm (mykmeans.jl) to seach for clusters in the generated dataset X

```julia
include("mykmeans.jl")
k = 3;    # The suggested number of clusters (you should also repeatedly try k = 2, 4 and 5)
Cid, CS, J = mykmeans(X, k);
```

### We plot the solution with coloring of each identified cluster

```julia
# Define and display a plot of the clustered data and the cluster centers(CS):
p = plot(title = "Clustering example with random generated 2D data", 
    label = " ", legend = :bottomright, size = (600, 400))
for i=1:k
    snr = vec(Cid.==i) # the sample numbers of the j-th cluster
    scatter!(p, X[snr,1], X[snr,2], label = string("Cluster ",i))
end
scatter!(p, CS[:,1],CS[:,2], marker = :star, markersize = 8, color = :orange, label = "CS") 
display(p)
```

### Plotting the objective function values for the iteratrive clustering process

```julia
# Define and display a plot of the objective function values reflecting the clustering process
Jp =plot(J, linestyle = :dashdot, title = "Objective function (J) values - monitoring the clustering process", 
    ylabel = "J (the mean squared distace)", xlabel = "Interations", label = "J")
display(Jp)
```

## A random "artificial" dataset X containing 3 point-clouds in 3 dimensions:

```julia
## Visualization for 3D data
# Here is a corresponding 3-dimensional dataset:
Random.seed!(1234) # This line assures the same random dataset to be generated each time.
nn = 100; 
X = [randn(nn,3)*0.7  .+ [1  1  1];
     randn(nn,3)*0.7  .+ [1 -2 -0];
     randn(nn,3)*0.7  .+ [0  0 -3]];
```

```julia
# Define and display a plot of the raw random data
sp = scatter(X[:,1],X[:,2],X[:,3], title = "Random pointcloud 3d-data", legend = false, camera = (75,10))
display(sp)
```

### We use our k-means algorithm to cluster the 3d-data. 

```julia
k = 3;    # The suggested number of clusters (you should also repeatedly try k = 3, 4 and 5)
Cid, CS, J = mykmeans(X, k);
```

```julia
##
# Plotting the clustered data and the cluster-centers:
colors = [:red, :blue, :green, :cyan, :magenta, :black];

# Define and display a plot of the clustered data and the cluster centers(CS):
p = plot(legend = :bottomright, title = "Clustering example with random generated 3D data", size = (600,400))
for j = 1:k
    snr = vec(Cid.==j); # the sample numbers of the j-th cluster
    # Plot only this group:
    scatter!(p, X[snr,1],X[snr,2],X[snr,3],color = colors[j], markersize = 5, label = string("Group ", j))
end
scatter!(p, CS[:,1],CS[:,2], CS[:,3], marker = :star, markersize = 12, color = :orange, label = "CS", camera = (75,10)) # Plotting the cluster centers
display(p)
```

### Plotting the objective function values for the iteratrive clustering process

```julia
# Define and display a plot of the objective function values reflecting the clustering process
Jp = plot(J, linestyle = :dashdot, title = "Objective function (J) values - monitoring the clustering process", 
    ylabel = "J (the mean squared distace)", xlabel="Interation", label="J")
display(Jp)
```
